```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image1.PNG')
```

```python
plt.imshow(img)
plt.show()
```

```python
img_cvt = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cvt)
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.show()
```
